<h4> Pochodna z definicji

```python
def pochodna(f, h = 0.0001):
    return lambda arg: (f(arg+h)-f(arg))/h

df = pochodna(lambda x: x**2)      
print(df(4))
```

```python
x = np.linspace(-5,5,100)
y = x**2*np.sin(x)                                  #funkcja
pochodna_analityczna = 2*x*np.sin(x)+x**2*np.cos(x) #policzona analitycznie
df = pochodna(lambda x: x**2*np.sin(x), 0.001) #pochodna obliczona z definicji
pochodna_przyblizona = np.array([df(i) for i in x]) 

plt.plot(x,y, c="blue")
plt.plot(x,pochodna_analityczna, c="green")
plt.plot(x,pochodna_przyblizona, c="red")

plt.show()
```

```python
from math import factorial

def sinus_taylor(x):
    y = 0
    n = 1
    for i in range(1,20,2):
        y += n*x**(i)/factorial(i)
        n = n*(-1)
    return y
```

<h4> Twierdzenie Poissona

$\binom{n}{k}p^k(1-p)^{n-k} \rightarrow e^{-\lambda}\cdot \frac{\lambda^k}{k!}$, gdzie $\lambda \approx p\cdot n$.

```python
from math import comb, e, factorial

#rozklad dwupunktowy
def Binom(p,n,k):
    return comb(n,k)*(p**k)*((1-p)**(n-k))

#rozklad Poissona
def Poisson(lam, k):
    return e**(-lam) * lam**k / factorial(k)

#Jeżeli p - małe, i p*n stała liczba (gdy liczymy granice), to przyjmując lambda = p*n rozklady dwupunktowe i Poissona sa zblizone
Binom(1/100, 100, 3), Poisson(1/100*100, 3)
```

<h4> Regresja liniowa

```python
import matplotlib.pyplot as plt
import numpy as np


#Model regresji
def l_m(x,y):
    a = (np.mean(x)*np.mean(y) - np.mean(x*y))/((np.mean(x))**2-np.mean(x*x))
    b = np.mean(y) - np.mean(x)*a
    return (a,b)

#Wartosci y wyznaczone z modelu
def y_z_modelu(x,y):
    wsp = l_m(x,y)
    y_model = wsp[0]*x+wsp[1]
    return y_model

#Wsp. korelacji
def r2(x,y):
    r2 = np.cov(x,y)[0][1]/(np.std(x,ddof=1)*np.std(y,ddof=1))
    return r2
```

```python
x1 = np.array([1,2,3,4,5])
y1 = np.array([4.1,7.6,11.3,12,14.8])

plt.scatter(x1, y1)
plt.plot(x1, y_z_modelu(x1,y1), c="red")
plt.show()
```

```python
print(r2(x1,y1))
```

<h2> Szeregi czasowe

```python
#Rysujemy Yn = aYn-1 +bYn-2 + e, gdzie Y1 = 0, Y2 = -1 <--- model AR(2)

import matplotlib.pyplot as plt
import numpy as np

Y = np.array([0,-1])
a = 0.5
b = 0.3

#Genruje 100 wyrazow z tego ciągu
for i in range(1000):
    Y_n = a*Y[-1]+b*Y[-2] + np.random.uniform(-1,1,1)
    Y = np.append(Y,Y_n)

#Moge nayrsowac wykres zaleznosci n vs Y_n
plt.plot(Y)
plt.xlabel("n")
plt.ylabel("Y_n")
plt.show()
```

```python
from statsmodels.graphics import tsaplots
import matplotlib.pyplot as plt

#Rysujemy funkcje autokorelacji
fig = tsaplots.plot_acf(Y, lags=10) #wsp korelacji dla biezacej chwili i 10 wczesniejszych
plt.show()
```

```python
r2(Y[1:],Y[:-1])
```

```python
#Zaleznosc tylko od poprzedniej chwili

import matplotlib.pyplot as plt
import numpy as np

Y = np.array([2])
a = 1

#Genruje 100 wyrazow z tego ciągu
for i in range(1000):
    Y_n = a*Y[-1] + np.random.uniform(-1,1,1)/10
    Y = np.append(Y,Y_n)

#Moge nayrsowac wykres zaleznosci n vs Y_n
plt.plot(Y)
plt.xlabel("n")
plt.ylabel("Y_n")
plt.show()

from statsmodels.graphics import tsaplots
import matplotlib.pyplot as plt

#plot autocorrelation function
fig = tsaplots.plot_acf(Y, lags=10) #wsp korelacji dla chwili i 10 wczesniejszych
plt.show()
```
